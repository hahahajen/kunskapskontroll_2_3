### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Både via pyplot eller objektorienterat. Objektorienterat är i regel att föredra.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure: hela ritbordet och innehåller en eller flera axes.
    
Axes: Är en graf där data skrivs ut. En eller flera kan användas samtidigt i en Figure.

Axis: det är x eller y axeln i en graf. Kan även innehålla en z-axel.

Artist: Objekt som ritats ut i figuren. Linjer, punkter, labels etc
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Numpy arrays är att föredra. Exempelvis Data objects från Pandas fungerar också, men helst Numpy arrays.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# OO approach
x = np.linspace(-4, 4, 111)
y = x**2

fix, ax = plt.subplots()

ax.set_title('OO approach')
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.plot(x, y)
```

```python
# Pyplot approach
x = np.linspace(-4, 4, 111)
y = x**2

plt.title('Pyplot approach')
plt.xlabel('x')
plt.ylabel('y')

plt.plot(x,y)
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, axs = plt.subplots(1,2, figsize=(12, 4))

# scatter plot
axs[0].scatter(random_data_x, random_data_y, s=5)
axs[0].scatter(x, y, s=5)
axs[0].set_title('Scatter plot')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

# bar plot
axs[1].bar(names, values)
axs[1].set_title('Bar plot')
axs[1].set_xlabel('Fruit name')
axs[1].set_ylabel('Quantity')
```
